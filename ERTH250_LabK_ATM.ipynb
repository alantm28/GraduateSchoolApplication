```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import scipy.stats as st
```

```python
eq=np.loadtxt('quakes.txt')
quakes=eq.T
x=np.array([quakes[0]])
y=np.array([quakes[1]])
z=-1*np.array([quakes[2]])
fig = plt.figure() 
ax = plt.axes(projection="3d")
ax.scatter3D(x, y, z, marker = '+')
plt.title("Earthquake distribution by latitude, longitude, depth")
ax.set_xlabel('longitude') 
ax.set_ylabel('latitude') 
ax.set_zlabel('depth in km')
ax.view_init(elev=34,azim=135)
#I think that this viewpoint best illustrates the depth that earthquakes can occur at.
```
